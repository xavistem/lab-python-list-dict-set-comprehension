````python
# Lista de productos
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# 1. Inicializar inventario con dict comprehension
def initialize_inventory(products):
    return {
        product: int(input(f"Enter the quantity of {product}s available: "))
        for product in products
    }

# 2. Pedir número de pedidos y nombres con set comprehension
def get_customer_orders():
    n = int(input("Enter the number of customer orders: "))
    return {
        input("Enter the name of a product that a customer wants to order: ")
        for _ in range(n)
    }

# 3. Calcular estadísticas (igual que antes)
def calculate_order_statistics(customer_orders, products):
    total = len(customer_orders)
    percent = total * 100 / len(products)
    return total, percent

def print_order_statistics(stats):
    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {stats[0]}")
    print(f"Percentage of Unique Products Ordered: {stats[1]}")

# 4. Actualizar inventario y eliminar ceros con dict comprehensions
def update_inventory(customer_orders, inventory):
    # Restar 1 en los pedidos
    inv = {
        p: (inventory[p] - 1) if p in customer_orders else inventory[p]
        for p in inventory
    }
    # Quitar los que quedan a 0
    return {p: q for p, q in inv.items() if q > 0}

# 5. Calcular precio total usando sum con generator comprehension
# No devuelve una lista o  diccionario (no recuerdo), al estar hecho por IA, preguntar por qué
def calculate_total_price(customer_orders):
    return sum(
        float(input(f"Enter the price of {p}: "))
        for p in customer_orders
    )

# 6. Imprimir inventario actualizado con list comprehension + join
def print_updated_inventory(inventory):
    print("\nUpdated Inventory:")
    print(
        "\n".join(
            f"{p}: {q}" for p, q in inventory.items()
        )
    )

# Programa principal
inventory = initialize_inventory(products)
customer_orders = get_customer_orders()
order_stats = calculate_order_statistics(customer_orders, products)
print_order_statistics(order_stats)
inventory = update_inventory(customer_orders, inventory)
print_updated_inventory(inventory)
total_price = calculate_total_price(customer_orders)
print(f"\nTotal Price: {total_price}")
````
